```python
import requests
import pandas as pd
import datetime
```

```python
# Ипользуем api_vk для парсинга метод wall.get

response = requests.get('https://api.vk.com/method/wall.get',
                        params={'access_token':'373a7bad373a7bad373a7bad33342f64a03373a373a7bad53f8fc5c14cc20b1e30a3c67',
                                'v': 5.131,
                                'domain':'soloviuha',
                                'count':100
                     }
                        )

data = response.json()
```

```python
# Создаем пустые списки для последующего добавления в них данных

likes=[]  # кол-во лайков
dates=[]  # даты
id=[]     # Id-номер поста



# С помощью цикла заполним списки

tz = datetime.timezone(+datetime.timedelta(hours=3))  # выбор часового пояса (GMT+3) - Москва

for post in data['response']['items']:
  likes.append(post['likes']['count']) # добавление лайков

  date_ = datetime.datetime.fromtimestamp(post['date'],tz).strftime('%Y-%m-%d %H:%M:%S') # перевод даты из UNIX
  dates.append(date_)  # добавление дат

  id.append(post['id'])  # добавление ID-номеров
```

```python
# Преобразование данных в таблицу

df = pd.DataFrame({
    'post_id': id,
    'date': dates,
    'likes': likes
})
```
